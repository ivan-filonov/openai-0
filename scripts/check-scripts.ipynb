```python
%matplotlib inline
from IPython.display import clear_output, Image as Jmage, display
import os
from PIL import Image
import signal
import subprocess
import sys
import time


def display_image(path):
    try:
        display(Image.open(path))
    except:
        print(f'no image: {path}')

def probe_state(app_key, state_png_path, clear=False):
    try:
        subprocess.run(['pkill', '-HUP', '-f', app_key], check=False)
        time.sleep(1)
        if clear:
            clear_output(False)
    except:
        pass
    display_image(state_png_path)
```
